## RANDOM FOREST REGRESSOR

```python
# Import necessary libraries for modeling
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Prepare features (X) and target (y) for training
X_train = train_state.drop('Reservoir', axis=1)  # Drop the 'Reservoir' column since it's an identifier
y_train = train_production.drop('Reservoir', axis=1)  # Drop the 'Reservoir' column in production data
```

```python
# Initialize and train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

## VALIDATION

```python
# Predicting on the validation set
X_val = val_state.drop('Reservoir', axis=1)  # Prepare validation features
y_val = val_production.drop('Reservoir', axis=1)  # Prepare validation targets
```

```python
print("Training Features:")
print(X_train.columns)
print("\nValidation Features:")
print(X_val.columns)

```

```python
from sklearn.metrics import r2_score
import numpy as np

# Predict on the validation set and evaluate
y_val_pred = model.predict(X_val)
val_mse = mean_squared_error(y_val, y_val_pred)
print(f'Validation Mean Squared Error: {val_mse}')
val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
print(f'Validation Root Mean Squared Error: {val_rmse}')
val_r2 = r2_score(y_val, y_val_pred)
print(f'Validation R^2 Score: {val_r2}')

# Predicting on the test set (state data only)
X_test = test_state.drop('Reservoir', axis=1)  # Ensure test data has similar features as train data
print("\nTest Features:")
print(X_test.columns)
test_predictions = model.predict(X_test)
```

```python

# Prepare the prediction output
predicted_production = pd.DataFrame(test_predictions, columns=['Gas production cumulative', 'Oil production cumulative', 'Water production cumulative'])
predicted_production['Reservoir'] = test_state['Reservoir']
print(predicted_production.head())

# Save predictions to CSV (optional)
predicted_production.to_csv(r'C:\Users\DANIELRAZAK\OneDrive\Desktop\VSCODE\EXXON MOBIL\student_starter_pack\dataset\AVERAGE\Reservoir_production.csv', index=False)
```

# Feature Scaling using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Scale the training and validation features
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

```

# Hyperparameter Tuning using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(RandomForestRegressor(random_state=42), param_distributions=param_grid, 
                                   n_iter=10, cv=5, n_jobs=-1, random_state=42, scoring='neg_mean_squared_error')

# Fit RandomizedSearchCV on the scaled training data
random_search.fit(X_train_scaled, y_train)

# Best model after tuning
best_model = random_search.best_estimator_

# 3. Train the optimized model
best_model.fit(X_train_scaled, y_train)

# 4. Predict on the scaled validation set
y_val_pred = best_model.predict(X_val_scaled)

# 5. Evaluate the performance
val_mse = mean_squared_error(y_val, y_val_pred)
val_rmse = np.sqrt(val_mse)
val_r2 = r2_score(y_val, y_val_pred)

print(f'Validation Mean Squared Error: {val_mse}')
print(f'Validation Root Mean Squared Error: {val_rmse}')
print(f'Validation R^2 Score: {val_r2}')
```
